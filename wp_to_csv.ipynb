```python
import pandas as pd
```

```python
df = pd.read_excel(r'/content/drive/MyDrive/eng/Robin.xlsx')
df[['talker', 'message']] = df[[' talker', ' message']]
df = df[['talker','message']]
df=df.astype(str)
df
```

```python
df['Combined_Message'] = df.groupby(df['talker'].ne(df['talker'].shift()).cumsum())['message'].transform(lambda x: ', '.join(x))

# Drop duplicate rows
df = df.drop_duplicates(subset=['talker', 'Combined_Message']).reset_index(drop=True)
df = df[['talker','Combined_Message']]
df = df.drop(index = 0 ,axis =1)
df
```

```python
response = []
message = []
for i in range(len(df)):
  if df['talker'].iloc[i] == 'Alihan Sağöz':
    
    response.append(df['Combined_Message'].iloc[i])
  else:
    message.append(df['Combined_Message'].iloc[i])
message
```

```python
contexted = []
n = 7
for i in range(n, len(df['Combined_Message'])):
  row = []
  prev = i - 1 - n # we additionally subtract 1, so row will contain current response and 7 previous responses  
  for j in range(i, prev, -1):
    row.append(df['Combined_Message'][j])
  contexted.append(row)
columns = ['response', 'context'] 
columns = columns + ['context/'+str(i) for i in range(n-1)]
df2 = pd.DataFrame.from_records(contexted, columns=columns)
df2
```

```python
df2.to_csv(r'/content/drive/MyDrive/eng_ready/Robin.csv')
```
